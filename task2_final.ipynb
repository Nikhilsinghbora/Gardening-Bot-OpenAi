```python
def collect_messages(_):
  prompt = inp.value_input
  inp.value = ''
  pinecone_reply = index.similarity_search_with_score(query = prompt,k = 3)
  context = ""
  for i in pinecone_reply:
    if i[1] >= 0.75:
      context += i[0].page_content

  template = "You are a gardening assistant bot, an automated service that answers the queries of the people \
    interested in gardening. Greet the user first in a friendly manner and then answer the question as truthfully as possible \
    using the provided context, \
    and if the answer is not contained within the text and requires some latest information to be updated, print \
    'Sorry Not Sufficient context to answer query' \
    The context is as follows:\
    {context}"
  prompt_template = PromptTemplate.from_template(template)
  prompt_message = prompt_template.format(context = context)
  messages.append(SystemMessage(content = prompt_message))
  messages.append(HumanMessage(content = prompt))
  reply = chat(messages)

  panels.append(
      pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
  panels.append(
      pn.Row('Assistant:', pn.pane.Markdown(reply.content, width=1000, styles={'background-color': '#F6F6F6'})))

  return pn.Column(*panels)
```

```python
messages = []
import panel as pn  # GUI
pn.extension()
panels = []
inp = pn.widgets.TextInput(value="Hi", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="Chat !",button_type='primary')

interactive_conversation = pn.bind(collect_messages, button_conversation)

dashboard = pn.Column(
    inp,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True),
)

dashboard
```
